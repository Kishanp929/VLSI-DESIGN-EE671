```python
N = [ 1 , 2 , 3  , 4 , 5 , 6 , 7 , 8]

delay = [  40.01492 , 52.58112 , 64.78109
          , 76.88766 , 88.96189 , 101.0421
           , 113.1355 , 125.2437 ]        # in ps

import matplotlib.pyplot as plt

# Sample data
# Create the plot
plt.plot(N, delay, label='X and Delay ', color='black', linestyle='-.')

# Add labels and title
plt.xlabel('No of Inverter for Load')
plt.ylabel('Delay (in ps)')
plt.title('Graph to find parasitic delay and Template inverter Delay')
plt.grid(True)  # Optional: Add grid lines
# Show the plot
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([ 1 , 2 , 3  , 4 , 5 , 6 , 7 , 8]).reshape(-1, 1)  # Reshape to a 2D array
y = np.array([  40.01492 , 52.58112 , 64.78109
          , 76.88766 , 88.96189 , 101.0421
           , 113.1355 , 125.2437 ] )

# Create a Linear Regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Get the slope (coefficient) and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Print the results
print(f"Slope (Coefficient): {slope}")
print(f"Intercept: {intercept}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for two lines
            # Second line (cos(x))

# Create a plot
plt.figure(figsize=(8, 6))  # Optional: Set the figure size

# Plot the first line in blue
plt.plot(X, delay, label='X and Delay ',linewidth=5 , color='black', linestyle='-')

# Plot the second line in red with markers
Ynew = slope*X + intercept
plt.plot(X, Ynew, label='Linearly fitted model', color='red' , linestyle='-')

# Add labels and a legend
plt.xlabel('No of inverter on Load side')
plt.ylabel('Delay (in ps)')
plt.title('Showing Linearly Fitted Model')
plt.legend()

# Show the plot
plt.grid(True)  # Optional: Add grid lines
plt.show()
```
